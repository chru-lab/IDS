# k-means / k-Mitten
Beispiel: Klick-Rate (Datei 'clicks.txt')
- der Einfachheit halber wieder in 2D (d.h. Daten mit nur zwei Attributen)
- Ausgangslage: Nutzerverhalten bzgl. angezeigter Werbung (x) und Anzahl Klicks (y)
- Darstellung als Streudiagramm
- 360 beliebige Nutzer*innen

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors
```

```python
data = np.loadtxt("clicks.txt", delimiter=",")
data
```

```python
plt.plot(data[:,0], data[:,1], 'r.')
```

```python
# Algorithmus importieren
from sklearn.cluster import KMeans
# Medell erzeugen (default=8, iterationsgrenze 300 )
model = KMeans(n_clusters=3, max_iter=300)
# klassifiziere mir meine Daten 
y_pred = model.fit_predict(data)
```

```python
# hat für jedes Element eine Klasse vorgeschlagen
y_pred
```

```python
plt.axis([0,70,0,10])
colors = ['red', 'blue', 'green']
cmap = matplotlib.colors.ListedColormap(colors)
plt.scatter(data[:,0], data[:,1], c=y_pred, cmap=cmap)
#plt.scatter(data[:,0], data[:,1], c=y_pred)
plt.show()
```

```python
# das gleiche mit 4 CLustern
model = KMeans(n_clusters=4, max_iter=300) 
y_pred = model.fit_predict(data)
plt.axis([0,70,0,10])
plt.scatter(data[:,0], data[:,1], c=y_pred)
plt.show()
```

```python
model.inertia_
```

## Ellenbogenmethode
- k-Mitten versucht, die Trägheit (inertia) bzgl. der Mittelpunkte zu minimieren
- Trägheit des Modells: model.inertia_
- guter k-Wert am Knick des Ellenbogens

```python
inert = []
for k in range(1,11):
    model = KMeans(n_clusters=k)
    model.fit(data)
    inert.append(model.inertia_)
```

```python
inert
```

```python
# Problem x-werte Fehlen -> k-werte nehmen. D.h. numerieren
# linspace erzeugt eine ANzahl werte zwischen zwei ziffern, welche den gleichen abstand haben  
# x = np.linspace(0,1,50)
x = np.linspace(1,10,10)
```

```python
plt.plot(x, inert, 'bo')
plt.plot(x, inert, 'b-')
plt.show()
```

Optimaler wert ist in diesem Plot k = 3 

## Mit Testdaten aus Scikit-learn arbeiten 

```python
from sklearn.datasets import make_moons
data, y_true = make_moons(n_samples=300, noise=0.05)
plt.scatter(data[:,0],data[:,1], c=y_true) # c = classification
```

```python
# Date modellieren
model = KMeans(n_clusters=4, max_iter=300) 
y_pred = model.fit_predict(data)
#plt.axis([0,70,0,10])
plt.scatter(data[:,0], data[:,1], c=y_pred)
plt.show()
```

```python
inert = []
for k in range(1,11):
    model = KMeans(n_clusters=k)
    model.fit(data)
    inert.append(model.inertia_)
```

```python
x = np.linspace(1,10,10)
```

```python
plt.grid(True)
plt.plot(x, inert, 'bo')
plt.plot(x, inert, 'b-')
plt.show()
```
