```python
# Daten einlesen
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt("credit.txt", delimiter=",")
print(data)
```

```python
# Daten Plotten 
plt.axis([10,80,0,120])
colors = ['red', 'blue'] # Array mit den beiden farben 
# Farbcodierungen in Colormap speichern 
import matplotlib.colors 
cmap = matplotlib.colors.ListedColormap(colors)
plt.scatter(data[:,0], data[:,1], c=data[:,2], cmap=cmap)
plt.show()
cmap
```

```python
# mit Legende Plotten
scatter = plt.scatter(data[:,0], data[:,1], c=data[:,2], cmap=cmap)
handles, labels = scatter.legend_elements()
labels = ['niedrig', 'hoch']
plt.legend(handles, labels, frameon=False, loc='upper left')
plt.show()
```
